PREPROCESSING
1. RESIZING
2. FLATTENING

```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

FLATTENING: converting matrix into vector

```python
#EG:
a=np.array([[1,2,3],
            [4,5,6]])
a.flatten()
```

```python
target=[]
images=[]
flat_data=[]
DATADIR='/content/images1'
CATAGORIES=['Bhuvneswar Kumar','MS Dhoni','PV Sindhu','Siraj','suryakumar yadav']
for category in CATAGORIES:
   class_num =CATAGORIES.index(category)#giving index numbers
   path=os.path.join(DATADIR,category)# showing path of the image
   print(path)
   for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    #print(img_array) # shows the matrix form
    #print(img_array.shape) # gives height, weight and width
   # plt.imshow(img_array)
    #break
    #RESIZING
    img_resized=resize(img_array,(150,150,3))
    #flattening
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)    
```

```python
len(flat_data[0]) #in one single dimension
```

```python
unique,count=np.unique(target,return_counts=True)
plt.bar(CATAGORIES,count)
```

```python
# split data into training and testing
```

```python
# sklearn - builds the model
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(flat_data,target,
                                  test_size=0.3,random_state=109)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid=[
    {'C':[1,10,100,1000],'kernel':['linear']},
    {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']}
]

svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
# save the model using pickle library
import pickle
pickle.dump(clf,open('img_model.p','wb'))
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import pickle
from PIL import Image
st.title('Image Classifier using machine learning')
st.text('upload the image')
model = pickle.load(open('img_model.p','rb'))
uploaded_file = st.file_uploader("choose an image....", type="jpg")
if uploaded_file is not None:
  img = Image.open(uploaded_file)
  st.image(img,caption='Uploaded Image')

  if st.button('PREDICT'):
     CATAGORIES=['Bhuvneswar Kumar','MS Dhoni','PV Sindhu','Siraj','suryakumar yadav']
     st.write('result.....')
     flat_data=[]
     img = np.array(img)
     img_resized = resize(img,(150,150,3))
     flat_data.append(img_resized.flatten())
     flat_data = np.array(flat_data)
     y_out = model.predict(flat_data)
     y_out = CATAGORIES[y_out[0]]
     st.title(f'PREDICTED OUTPUT: {y_out}')
     q= model.predict_proba(flat_data)
     for index, item in enumerate(CATAGORIES):
       st.write(f'{item: {q[0][index]*100}')
```
